## Desvio Padrão:

O desvio padrão pode ser utilizado para ver a dispersão entre os valores de um mesmo grupo, em cada categoria. Assim, é possível comparar os valores entre os grupos, para ver o quanto variam os valores entre os países, e também verificar se uma análise estatística utilizando a média/mediana pode ou não ser viável, considerando que se o desvio padrão de um grupo for muito maior do que o outro, ou mesmo muito grande relativo à sua média, uma análise deste tipo pode ser focada somente em alguns países que compõe a média, ao que os outros países estão dispersos. 

```python
x=pd.pivot_table(data=data, index= 'UE', aggfunc=np.std)
x=x[['2010_Exports','2014_Exports','2017_Exports','2010_PIB','2014_PIB','2017_PIB','2010_FDI','2014_FDI','2017_FDI','2010_IDH','2013_IDH','2015_IDH',]]
x
```

Quando se trata das exportações, a tabela acima indica que os países europeus que não participam da União Européia têm o menor desvio padrão, seguido pelos não europeus, e depois pelos participantes. Entretanto, vale ressaltar que os países participantes têm uma média elevada com relação às outras, e devido à isso, um desvio padrão mais alto não necessariamente significa que os resultados estão mais dispersos entre si, e sim que eles potencialmente se dispersam de maneira similar, uma vez que a magnitude com qual os valores dos países participantes mudam é maior também.

Já com relação ao PIB, os europeus que não participam da UE têm o maior desvio padrão, seguidos pelos participantes e depois pelos não europeus. Entretanto, a mesma justificativa que a do índicador de exportação não pode ser inteiramente utilizada. Enquanto, de fato, o menor desvio padrão pertence ao grupo com a menor média, o do médio pertence ao grupo com a maior média. Isso pode indicar que de fato existe uma maior dispersão entre os dados dos europeus não participantes, algo que poderia sugerir que uma análise baseada na média pode não relatar a situação geral da maioria dos países, e sim somente daqueles que estão na média. No entanto, nas células da análise de cada indicador, será conduzida uma comparação entre a média e a mediana em um histograma, para cada grupo e para cada ano, fazendo com que a validação por trás do uso da média seja obtida por outra análise, sem ser o desvio padrão.

No que tange aos investimentos estrangeiros que entram nos países, a mesma tendência das exportações pode ser evidenciada com relação à ordem dos grupos. Como nas exportações, os países europeus participantes têm a maior média para acompanhar o seu maior desvio padrão, assim justificando o valor elevado. Entretanto, vale ressaltar que ao longo dos três anos, o desvio padrão dos participantes e dos não europeus diminui significantemente, mas como justificativa (de certa forma) a sua média diminuiu de acordo.

E por fim, o desvio padrão do Índice de Desenvolvimento Humano se mostrou muito menor nos países que participam da UE, seguido pelos países europeus não participantes, e por fim os não europeus. Esses valores não variaram muito ao longo do tempo, acompanhando a estabilidade da média ao longo dos anos analisado. Vale ressaltar que ao comparar com a média, os países não europeus apresentam uma disperção muito maior entre os países, enquanto os que participam da UE se mantiveram muito mais próximos, e os europeus não participantes apresentaram um dispersão maior que a dos países da UE, mas consideravelmente menor que a dos não europeus.

## PIB:

Para toda variável, será utilizado um grupo de histogramas para analisar, em especial, a média e a mediana. Serão utilizados nove histogramas por indicador, três para cada grupo (um para cada ano). Abaixo, estão os gráficos referentes ao PIB per capita ajustado à inflação.

```python
def verticalLine(plot, x, y, color, nome):
    plot.axvline(x, color=color, linestyle="dashed", linewidth=2)
    plot.text(1.01*x, y, nome, color=color) 
fig = plt.figure(figsize=(15, 15))
def plots(grupo,coluna,nome,plot,cor,pos1,pos2):
    plt.subplot(plot)
    plot = grupo[coluna].plot.hist(bins=10, title='{0}: {1}'.format(nome, coluna), alpha=0.9, color=cor)
    plt.ylabel('Frequência Absoluta')
    plt.xlabel('PIB (ajustado à inflação) em USD')
    plot.set_xlim([0,110000])
    plt.tight_layout()
    verticalLine(plot, grupo[coluna].mean(), pos1, "purple", "Média")
    verticalLine(plot, grupo[coluna].median(), pos2, "black", "Mediana")

plots(uniao,'2010_PIB','UE',431,'b',6,5)
plots(uniao,'2014_PIB','UE',432,'b',6,5)
plots(uniao,'2017_PIB','UE',433,'b',6,5)
plots(euro,'2010_PIB','Europeus não participantes',434,'r',4,3)
plots(euro,'2014_PIB','Europeus não participantes',435,'r',4,3)
plots(euro,'2017_PIB','Europeus não participantes',436,'r',4,3)
plots(ne,'2010_PIB','Não Europeus',437,'g',55,50)
plots(ne,'2014_PIB','Não Europeus',438,'g',55,50)
plots(ne,'2017_PIB','Não Europeus',439,'g',55,50)
```

Um fator comum entre os gráficos acima, é o fato de que a média está sempre à direita da mediana, e assim mostram uma assimetria direita. Isso significa que os poucos países que apresentam um índice muito elevado com relação aos outros, puxam a média em direção à eles, quebrando uma possível simetria entre os dados, ao que a média e a mediana ficam diferentes. Entretanto, em nenhum dos gráficos a diferença entre a média e a mediana é grande o suficiente para invalidar a utilização da média para a análise. Vale ressaltar que os dados dos europeus não participantes têm uma maior diferença entre a média e a mediana. Entretanto, isso é bastante devido ao fato de que esse grupo tem somente 12 países, e assim é muito mais fácil para um país com um PIB muito mais alto deslocar a média. Mas, para dar continuidade à análise, e considerando o fato de que a média ainda está perto da faixa com maior densidade de países, pode-se utilizar a média para a análise.

Outra análise dos histogramas mostra como que em todos os grupos, a grande maioria dos países estão concentrados na parte com menor PIB. Entretanto, os países da UE apresentam uma curva mais sutil ao que aumenta o valor do PIB, tendo assim mais países em faixas maiores, com relação aos outros grupos.



Abaixo, está o gráfico que será utilizado (em todos os indicadores) para comparar a média do PIB per capita de cada grupo ao longo dos anos. Esse gráfico pode fornecer uma maneira relativamente simples de observar a tendência dos dados durante os anos analisados, e assim chegar em conclusões sobre a efetividade de cada grupo com relação ao indicador em questão.

```python
media_GDP_UE_2010= uniao['2010_PIB'].mean()
media_GDP_Europeus_2010= euro['2010_PIB'].mean()
media_GDP_NE_2010= ne['2010_PIB'].mean()

media_GDP_UE_2014= uniao['2014_PIB'].mean()
media_GDP_Europeus_2014= euro['2014_PIB'].mean()
media_GDP_NE_2014= ne['2014_PIB'].mean()

media_GDP_UE_2017= uniao['2017_PIB'].mean()
media_GDP_Europeus_2017= euro['2017_PIB'].mean()
media_GDP_NE_2017= ne['2017_PIB'].mean()

media_GDP_UE = (media_GDP_UE_2010 + media_GDP_UE_2014 + media_GDP_UE_2017)/3
media_GDP_Europeus= (media_GDP_Europeus_2010 + media_GDP_Europeus_2014 + media_GDP_Europeus_2017)/3
media_GDP_NE = (media_GDP_NE_2010 + media_GDP_NE_2014 + media_GDP_NE_2017)/3

medias_gdp = {'União Européia': media_GDP_UE , 'Europeus não participantes': media_GDP_Europeus , 'Não Europeus':media_GDP_NE}
names=list(medias_gdp.keys())
values=list(medias_gdp.values()) 

plt.bar(names,values, color = ('blue','red','green'))
plt.title("PIB")
plt.ylabel('PIB (ajustado à inflação) em USD')
plt.show()
```

Acima, pode se identificar que os países da UE têm o maior PIB per capita, seguido pelos europeus não participantes, e depois pelos não europeus. Com relação aos países da União Européia, os europeus não participantes têm um PIB de em torno de 32% menor, enquanto o dos não europeus é de aproximadamente 66% menor. Dessa forma, a análise geral proposta por esse gráfico aponta ao fato de que estar na União Européia pode ser bastante favorável com relação ao PIB per capita, especialmente ao considerar a diferença entre os países europeus que participam e os que não participam da UE.


Já o gráfico abaixo mostra os pontos coletados e a linha de tendência com relação ao PIB de cada grupo analisado. Essa análise consegue explicar melhor a taxa de variação de cada grupo, e o seu comportamento ao longo do tempo, contrário ao gráfico acima que exibe a média durante o tempo analisado. Os dois gráficos, portanto, funcionam de maneira bastante complementar, ao que juntos é possível ter uma análise completa sobre o PIB de cada país.

```python
anos= [2010,2014,2017]
medias_gdp_UE=[media_GDP_UE_2010, media_GDP_UE_2014, media_GDP_UE_2017]
medias_gdp_Europeus=[media_GDP_Europeus_2010, media_GDP_Europeus_2014, media_GDP_Europeus_2017]
medias_gdp_NE=[media_GDP_NE_2010, media_GDP_NE_2014, media_GDP_NE_2017]

plt.scatter(anos,medias_gdp_UE, label = 'União Européia', color= 'b')
z = np.polyfit(anos, medias_gdp_UE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"b--")


plt.scatter(anos,medias_gdp_Europeus, label = 'Europeu não participante', color = 'r')
z = np.polyfit(anos, medias_gdp_Europeus, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"r--")


plt.scatter(anos,medias_gdp_NE, label = 'Não Europeu', color = 'g')
z = np.polyfit(anos, medias_gdp_NE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"g--")


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('PIB (ajustado à inflação) em USD')
plt.xlabel('Anos')
plt.title('PIB')
plt.grid(True)
plt.show()
```

O gráfico acima mostra como os países pertencentes à União Européia apresentam não somente uma magnitude dos valores do PIB per capita superior aos outros grupos, mas também apresenta uma taxa de variação maior, algo que mostra uma tendência melhor que os outros grupos. Já os países europeus não participantes estão piores que os da UE, mas melhores que os não europeus, tanto na taxa de variação, quanto na magnitude dos valores. Isso demonstra que com relação aos outros países no mundo, simplesmente ser um país europeu já pode significar que um país terá uma probabilidade mais alta de ter um PIB elevado. E o mesmo pode ser dito para os pa;ises que participam da UE, com relação aos europeus não participantes, uma vez que os países da UE apresentam tendências bastante superiores.

## Investimento Estrangeiro:

Investimento estrangeiro pode servir como um ótimo indicador da confiança de investidores em um país, algo que reflete sua situação econômica. Portanto, esse indicador será usado para comparar a situação econômica dos grupos. Abaixo estão os histogramas, da mesma forma, e destinado à mesma análise, que fora feita para o PIB, com os mesmos limites do eixo x, para uma comparação dos valores entre os grupos.

```python
fig = plt.figure(figsize=(15, 15))
def plots(grupo,coluna,nome,plot,cor,pos1,pos2):
    plt.subplot(plot)
    plot = grupo[coluna].plot.hist(bins=10, title='{0}: {1}'.format(nome, coluna), alpha=0.9, color=cor)
    plt.ylabel('Frequência Absoluta')
    plt.xlabel('Entrada de investimentos estrangeiros como % do PIB')
    plt.tight_layout()
    plot.set_xlim([0,115])
    verticalLine(plot, grupo[coluna].mean(), pos1, "b", "Média")
    verticalLine(plot, grupo[coluna].median(), pos2, "black", "Mediana")

plots(uniao,'2010_FDI','UE',331,'r',15,12)
plots(uniao,'2014_FDI','UE',332,'r',12,10)
plots(uniao,'2017_FDI','UE',333,'r',11,10)
plots(euro,'2010_FDI','Europeus não participantes',334,'g',4,3)
plots(euro,'2014_FDI','Europeus não participantes',335,'g',3.4,3)
plots(euro,'2017_FDI','Europeus não participantes',336,'g',6,5)
plots(ne,'2010_FDI','Não Europeus',337,'yellow',110,100)
plots(ne,'2014_FDI','Não Europeus',338,'yellow',50,40)
plots(ne,'2017_FDI','Não Europeus',339,'yellow',35,30)
```

Apesar de mostrar que os países da UE têm, em média, valores maiores de FDI, seguido pelos países não europeus, os gráficos acima estão de certa forma limitados com relação à análise que eles podem prover. Diferentemente dos outros índices, colocar o FDI no mesmo limite do eixo x entre os grupos faz com que muitos histogramas acima não possam ser visualizados facilmente. Portanto, abaixo estão os histogramas com os limites diferentes, para observar melhor cada histograma.

```python
fig = plt.figure(figsize=(15, 15))
def plots(grupo,coluna,nome,plot,cor,pos1,pos2):
    plt.subplot(plot)
    plot = grupo[coluna].plot.hist(bins=10, title='{0}: {1}'.format(nome, coluna), alpha=0.9, color=cor)
    plt.ylabel('Frequência Absoluta')
    plt.xlabel('Entrada de investimentos estrangeiros como % do PIB')
    plt.tight_layout()
    verticalLine(plot, grupo[coluna].mean(), pos1, "b", "Média")
    verticalLine(plot, grupo[coluna].median(), pos2, "black", "Mediana")

plots(uniao,'2010_FDI','UE',331,'r',15,12)
plots(uniao,'2014_FDI','UE',332,'r',10,12)
plots(uniao,'2017_FDI','UE',333,'r',10,9)
plots(euro,'2010_FDI','Europeus não participantes',334,'g',4,3)
plots(euro,'2014_FDI','Europeus não participantes',335,'g',3.5,3)
plots(euro,'2017_FDI','Europeus não participantes',336,'g',6,5)
plots(ne,'2010_FDI','Não Europeus',337,'yellow',110,100)
plots(ne,'2014_FDI','Não Europeus',338,'yellow',45,35)
plots(ne,'2017_FDI','Não Europeus',339,'yellow',35,25)
```

Ao analisar os histogramas acima, pode se evidenciar que a média e a mediana estão, na maioria das vezes, um tanto quanto similares, tornando possível a análise pela média. Ademais, uma grande diferença dos histogramas do PIB, é que nem todos tem uma assimetria direita, ou seja, a média nem sempre está à direita da mediana. Essa irregularidade (com relação ao resto dos gráficos) ocorre nos países europeus não participantes em 2017. Devido ao fato da diferença ser sutil, a análise não muda drasticamente, mas vale ressaltar a razão por qual isso aconteceu. Essa assimetria esquerda é devido ao movimento negativo de investimento no país, ou seja, existe mais capital saindo dos investimentos no grupo de países, do que entrando, algo que diminui bastante a média. Vale ressaltar também, que ao comparar os histogramas, pode se evidenciar uma média/mediana diminuindo ao longo do tempo, na maioria dos casos, algo que será analisado nessa seção.

```python
media_FDI_UE_2010= uniao['2010_FDI'].mean()
media_FDI_Europeus_2010= euro['2010_FDI'].mean()
media_FDI_NE_2010= ne['2010_FDI'].mean()

media_FDI_UE_2014= uniao['2014_FDI'].mean()
media_FDI_Europeus_2014= euro['2014_FDI'].mean()
media_FDI_NE_2014= ne['2014_FDI'].mean()

media_FDI_UE_2017= uniao['2017_FDI'].mean()
media_FDI_Europeus_2017= euro['2017_FDI'].mean()
media_FDI_NE_2017= ne['2017_FDI'].mean()

media_FDI_UE = (media_FDI_UE_2010 + media_FDI_UE_2014 + media_FDI_UE_2017)/3
media_FDI_Europeus= (media_FDI_Europeus_2010 + media_FDI_Europeus_2014 + media_FDI_Europeus_2017)/3
media_FDI_NE = (media_FDI_NE_2010 + media_FDI_NE_2014 + media_FDI_NE_2017)/3

medias_fdi = {'União Européia': media_FDI_UE , 'Europeus não participantes': media_FDI_Europeus , 'Não Europeus':media_FDI_NE}
names=list(medias_fdi.keys())
values=list(medias_fdi.values())

plt.bar(names,values, color = ('r','g','yellow'))
plt.title("FDI")
plt.ylabel('Entrada de investimentos estrangeiros como % do PIB')
plt.show()
```

O gráfico acima, das médias dos valores de cada grupo ao longo dos anos, mostra um fenômeno muito interessante. Enquanto os países da União Européia tem os maiores valores, os países europeus não participantes têm uma entrada de capital estrangeiro menor do que os países não europeus, mostrando uma tendência econômica diferente da do PIB. Entretanto, ambas as análises comprovam que, considerando a média ao longo do período analisado, a União Européia apresenta, em ao menos esses dois fatores, um ritmo econômico um tanto quanto melhor que os outros grupos, considerando que os europeus não participantes tiveram 55% de investimento estrangeiro a menos que os da UE, enquanto os não europeus tiveram 38% a menos.

```python
anos= [2010,2014,2017]
medias_fdi_UE=[media_FDI_UE_2010, media_FDI_UE_2014, media_FDI_UE_2017]
medias_fdi_Europeus=[media_FDI_Europeus_2010, media_FDI_Europeus_2014, media_FDI_Europeus_2017]
medias_fdi_NE=[media_FDI_NE_2010, media_FDI_NE_2014, media_FDI_NE_2017]

plt.scatter(anos,medias_fdi_UE, label = 'União Européia', color= 'r')
z = np.polyfit(anos, medias_fdi_UE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"r--")


plt.scatter(anos,medias_fdi_Europeus, label = 'Europeu não participante', color = 'g')
z = np.polyfit(anos, medias_fdi_Europeus, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"g--")


plt.scatter(anos,medias_fdi_NE, label = 'Não Europeu', color = 'y')
z = np.polyfit(anos, medias_fdi_NE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"y--")


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Entrada de investimentos estrangeiros como % do PIB')
plt.xlabel('Anos')
plt.title('FDI')
plt.grid(True)
plt.show()
```

Outra análise bastante interessante é possibilitada pelo gráfico acima. Enquanto na média durante o período investigado, a UE tinha o valor mais alto de FDI, o gráfico acima demonstra que, apesar da média ser mais elevada, a taxa de variação da União Européia é negativa (como todas) e é a menor entre os três grupos. Isso significa que, caso as taxas de variação vistas acima se mantenham, a média da União Européia ao longo do tempo possa ser inferior à dos outros grupos, mostrando um sinal econômico negativo para os países da UE. Entretanto, pode-se dizer que a diminuição do FDI como % do PIB não é algo negativo, pois pode significar que a economia está focando em atividades que dependem menos de capital estrangeiro, algo que não é necessariamente ruim. Mas, para essa análise, o FDI será utilizado como um indicador da confiança do investidor estrangeiro no grupo de países, e que portanto reflete a situação econômica dos países dentro do grupo.

## Exportação:

A quantidade de exportação de bens vindos de um país é um indicador que reflete a capacidade de uma economia 'produzir seus próprios frutos', além de representar uma fonte de capital potencialmente significante. Por essa razão, esse indicador será utilizado na análise do impacto econômico de pertencer à União Européia, uma análise que também deve levar em conta o fato de que esse pertencimento pode facilitar o comércio com os outros países da UE. Abaixo, estão os histogramas que demonstram os dados utilizados.

```python
fig = plt.figure(figsize=(15, 15))
def plots(grupo,coluna,nome,plot,cor,pos1,pos2):
    plt.subplot(plot)
    plot = grupo[coluna].plot.hist(bins=10, title='{0}: {1}'.format(nome, coluna), alpha=0.9, color=cor)
    plt.ylabel('Frequência Absoluta')
    plt.xlabel('Exportação como % do PIB')
    plt.tight_layout()
    plot.set_xlim([0,230])
    verticalLine(plot, grupo[coluna].mean(), pos1, "darkgray", "Média")
    verticalLine(plot, grupo[coluna].median(), pos2, "black", "Mediana")

plots(uniao,'2010_Exports','UE',331,'r',8,6.5)
plots(uniao,'2014_Exports','UE',332,'r',9,7)
plots(uniao,'2017_Exports','UE',333,'r',9,7)
plots(euro,'2010_Exports','Europeus não participantes',334,'b',2.5,2)
plots(euro,'2014_Exports','Europeus não participantes',335,'b',2.7,2.3)
plots(euro,'2017_Exports','Europeus não participantes',336,'b',1.7,1.3)
plots(ne,'2010_Exports','Não Europeus',337,'yellow',50,40)
plots(ne,'2014_Exports','Não Europeus',338,'yellow',50,40)
plots(ne,'2017_Exports','Não Europeus',339,'yellow',30,22)
```

Essas figuras mostram como os países participantes da UE têm um índice maior de exportação, ao que têm mais países deslocados à direita com relação aos outros. Já os países europeus não participantes aparentam estar de certa forma concentrados entre a tendência entre os países não europeus (deslocados à esquerda) e os países participantes, ao que eles não têm países na faixa 0-25, enquanto os não europeus têm, na maioria dos casos.

Como na análise do FDI, os gráficos acima, por estarem no mesmo limite do eixo x, não estão adequados para uma análise da média/mediana. Para isso, esses gráficos estão abaixo, com os limites adequados para cada gráfico, algo que compromete a análise visual entre os gráficos, mas permite uma análise individual melhor.

```python
fig = plt.figure(figsize=(15, 15))
def plots(grupo,coluna,nome,plot,cor,pos1,pos2):
    plt.subplot(plot)
    plot = grupo[coluna].plot.hist(bins=10, title='{0}: {1}'.format(nome, coluna), alpha=0.9, color=cor)
    plt.ylabel('Frequência Absoluta')
    plt.xlabel('Exportação como % do PIB')
    plt.tight_layout()
    verticalLine(plot, grupo[coluna].mean(), pos1, "darkgray", "Média")
    verticalLine(plot, grupo[coluna].median(), pos2, "black", "Mediana")

plots(uniao,'2010_Exports','UE',331,'r',7.5,6)
plots(uniao,'2014_Exports','UE',332,'r',7.5,6)
plots(uniao,'2017_Exports','UE',333,'r',7.5,6)
plots(euro,'2010_Exports','Europeus não participantes',334,'b',2.7,2.3)
plots(euro,'2014_Exports','Europeus não participantes',335,'b',2.7,2.3)
plots(euro,'2017_Exports','Europeus não participantes',336,'b',1.5,1.1)
plots(ne,'2010_Exports','Não Europeus',337,'yellow',50,40)
plots(ne,'2014_Exports','Não Europeus',338,'yellow',50,40)
plots(ne,'2017_Exports','Não Europeus',339,'yellow',35,27)
```

Como pode ser evidenciado nos gráficos acima, a média e a mediana são bastante próximas, como nos outros índices, e portanto a média pode ser utilizada para as análises abaixo. Ademais pode se evidenciar uma ligeira assimetria direita em todos os gráficos, mas vale trazer atenção ao gráfico dos europeus não participantes de 2017. Esse gráfico exibe picos de frequência através de vários pontos de exportação. Essa informação demonstra que entre todos os países que estão no grupo, eles estão bastante dispersos entre si, algo que também é verdade para os outros anos, mas não de maneira tão drástica.

```python
media_Exports_UE_2010= uniao['2010_Exports'].mean()
media_Exports_Europeus_2010= euro['2010_Exports'].mean()
media_Exports_NE_2010= ne['2010_Exports'].mean()

media_Exports_UE_2014= uniao['2014_Exports'].mean()
media_Exports_Europeus_2014= euro['2014_Exports'].mean()
media_Exports_NE_2014= ne['2014_Exports'].mean()

media_Exports_UE_2017= uniao['2017_Exports'].mean()
media_Exports_Europeus_2017= euro['2017_Exports'].mean()
media_Exports_NE_2017= ne['2017_Exports'].mean()

media_Exports_UE = (media_Exports_UE_2010 + media_Exports_UE_2014 + media_Exports_UE_2017)/3
media_Exports_Europeus= (media_Exports_Europeus_2010 + media_Exports_Europeus_2014 + media_Exports_Europeus_2017)/3
media_Exports_NE = (media_Exports_NE_2010 + media_Exports_NE_2014 + media_Exports_NE_2017)/3

medias_Exports = {'União Européia': media_Exports_UE , 'Europeus não participantes': media_Exports_Europeus , 'Não Europeus':media_Exports_NE}
names=list(medias_Exports.keys())
values=list(medias_Exports.values())

plt.bar(names,values, color = ('r','b','yellow'))
plt.title("Exportação")
plt.ylabel('Exportação como % do PIB')
plt.show()
```

Ao analisar o gráfico acima, que mostra a média de exportação de cada grupo ao longo dos anos analisados, pode-se concluir que os países da UE têm uma quantidade de exportação muito maior, relativo ao outros países e seus PIBs. Relativo ao grupo da UE, os europeus não participantes tem 36% menos exportações relativo ao PIB, enquanto os não europeus tem 43% a menos. Essa pequena diferença entre os europeus não participantes e os não europeus, evidencia que com relação à exportação os países da UE de fato tem uma "vantagem" econômica, algo que pode ser consequência da "coesão econômica entre os países participantes" mencionada nos objetivos da UE.

```python
anos= [2010,2014,2017]
medias_Exports_UE=[media_Exports_UE_2010, media_Exports_UE_2014, media_Exports_UE_2017]
medias_Exports_Europeus=[media_Exports_Europeus_2010, media_Exports_Europeus_2014, media_Exports_Europeus_2017]
medias_Exports_NE=[media_Exports_NE_2010, media_Exports_NE_2014, media_Exports_NE_2017]

plt.scatter(anos,medias_Exports_UE, label = 'União Européia', color= 'r')
z = np.polyfit(anos, medias_Exports_UE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"r--")


plt.scatter(anos,medias_Exports_Europeus, label = 'Europeu não participante', color = 'b')
z = np.polyfit(anos, medias_Exports_Europeus, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"b--")


plt.scatter(anos,medias_Exports_NE, label = 'Não Europeu', color = 'y')
z = np.polyfit(anos, medias_Exports_NE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"y--")


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Exportação como % do PIB')
plt.xlabel('Anos')
plt.title('Exportação')
plt.grid(True)
plt.show()
```

As linhas de tendência do gráfico acima são diferentes, de certa forma, das outras mostradas ao longo do projeto. Isso se deve ao fato de que nos outros índices, elas eram todas crescentes ou decrescentes. No caso da exportação, somente os grupos europeus apresentam uma taxa de variação positiva, além de serem numéricamente superiores aos países não europeus nesse quesito. Isso significa que o 'espaço' que as exportações ocupariam no PIB dos países não europeus vem caindo, em geral, especialmente considerando que em 2010 os níveis dos não europeus e dos europeus não participantes eram bastante similares, mas em 2017 a diferença deles já era em torno de 12% das exportações deste. Apesar da diferença entre os europeus e o resto dos países, vale ressaltar a magnitude elevada dos participantes da UE, contra os que não participam, uma vez que essa diferença é de em torno de 45% da as exportações dos europeus não participantes, algo que é possívelmente consequência dos objetivos da UE, como foi explicado acima.

## Índice de Desenvolvimento Humano:

O Índice de Desenvolvimento Humano leva em conta três índices: alfabetização adulta, expectativa de vida, e potencial de compra. Esses três fatores refletem tanto a situação social de um país, quanto a situação econômica, ao que um IDH alto é comumente encontrado em países com condições financeiras para atenderem aos requisitos. Os histogramas abaixo demonstram como um IDH alto é bastante comum entre os países da UE, mas isso não é a realidade de todos os outros países.

```python
fig = plt.figure(figsize=(15, 15))
def plots(grupo,coluna,nome,plot,cor,pos1,pos2):
    plt.subplot(plot)
    plot = grupo[coluna].plot.hist(bins=10, title='{0}: {1}'.format(nome, coluna), alpha=0.9, color=cor)
    plt.ylabel('Frequência Absoluta')
    plt.xlabel('Índice de Desenvolvimento Humano')
    plt.tight_layout()
    plot.set_xlim([0.3,1])
    verticalLine(plot, grupo[coluna].mean(), pos1, "r", "Média")
    verticalLine(plot, grupo[coluna].median(), pos2, "black", "Mediana")

plots(uniao,'2010_IDH','UE',331,'orange',5,4)
plots(uniao,'2013_IDH','UE',332,'orange',5,4)
plots(uniao,'2015_IDH','UE',333,'orange',5,4)
plots(euro,'2010_IDH','Europeus não participantes',334,'b',2.5,2)
plots(euro,'2013_IDH','Europeus não participantes',335,'b',2.5,2)
plots(euro,'2015_IDH','Europeus não participantes',336,'b',2.5,2)
plots(ne,'2010_IDH','Não Europeus',337,'g',23,19)
plots(ne,'2013_IDH','Não Europeus',338,'g',23,19)
plots(ne,'2015_IDH','Não Europeus',339,'g',23,19)
```

Os gráficos acima permitem identificar que os países da UE estão concentrados muito mais para a direita do eixo x, mostrando um IDH alto através dos anos. Já os países europeus não participantes apresentam uma tendência de IDH mais concentrada nos valores altos, mas de forma menos extrema que os países da UE, mas ainda superando fortemente os países não europeus, que têm valores altos de frequência absoluta em várias faixas, baixas e altas. Além disso, o fato de que os histogramas permanecem bastante similares ao longo dos anos analisados indica o quão completo é o índice, pois ele não é algo que varia facilmente, uma vez que os indicadores que o compõe são bastante densos e difíceis de mudar em um curto período. Assim, a utilização desse índice se diferencia dos outros, pois ele dificilmente estará sujeito aos potenciais malefícios de escolher anos particularmente afetados por algum fenômeno que poderia alterar a análise.

Para analisar melhor a média e a mediana, cada gráfico deve estar em seu próprio limite no eixo x, como estão os histogramas abaixo.

```python
fig = plt.figure(figsize=(15, 15))
def plots(grupo,coluna,nome,plot,cor,pos1,pos2):
    plt.subplot(plot)
    plot = grupo[coluna].plot.hist(bins=10, title='{0}: {1}'.format(nome, coluna), alpha=0.9, color=cor)
    plt.ylabel('Frequência Absoluta')
    plt.xlabel('Índice de Desenvolvimento Humano')
    plt.tight_layout()
    verticalLine(plot, grupo[coluna].mean(), pos1, "r", "Média")
    verticalLine(plot, grupo[coluna].median(), pos2, "black", "Mediana")

plots(uniao,'2010_IDH','UE',331,'orange',5.5,4.5)
plots(uniao,'2013_IDH','UE',332,'orange',5.5,4.5)
plots(uniao,'2015_IDH','UE',333,'orange',5.5,4.5)
plots(euro,'2010_IDH','Europeus não participantes',334,'b',2.5,2)
plots(euro,'2013_IDH','Europeus não participantes',335,'b',2.5,2)
plots(euro,'2015_IDH','Europeus não participantes',336,'b',2.5,2)
plots(ne,'2010_IDH','Não Europeus',337,'g',23,19)
plots(ne,'2013_IDH','Não Europeus',338,'g',23,19)
plots(ne,'2015_IDH','Não Europeus',339,'g',23,19)
```

Como na maioria dos outros histogramas utilizados no projeto, os países europeus não participantes têm a média depois da mediana, tendo assim uma assimetria direita, algo que indica que os países com um IDH muito desenvolvido conseguem mudar a média, apesar da grande parte dos países estarem em torno de 0.75-0.8. Vale ressaltar, no entanto, que esses valores em si já são moderadamente altos para o IDH. Em contrapartida, o grupo dos países não europeus e os da UE apresentam a média levemente antes da mediana, algo que indica uma assimetria esquerda, contrário ao fenômeno com os países europeus não participantes. Essa assimetria significa que aqueles países com um IDH relativamente baixo no grupo, abaixam a média até antes do país cujo valor representa 50% dos dados, justamente por serem significantemente menores que a mediana. Isso indica que há uma certa desigualdade com relação ao IDH nos grupos, mas, a média permanece sendo um bom valor para basear as análises, uma vez que ela está bastante perto da mediana nos dois casos.

```python
media_IDH_UE_2010= uniao['2010_IDH'].mean()
media_IDH_Europeus_2010= euro['2010_IDH'].mean()
media_IDH_NE_2010= ne['2010_IDH'].mean()

media_IDH_UE_2013= uniao['2013_IDH'].mean()
media_IDH_Europeus_2013= euro['2013_IDH'].mean()
media_IDH_NE_2013= ne['2013_IDH'].mean()

media_IDH_UE_2015= uniao['2015_IDH'].mean()
media_IDH_Europeus_2015= euro['2015_IDH'].mean()
media_IDH_NE_2015= ne['2015_IDH'].mean()

media_IDH_UE = (media_IDH_UE_2010 + media_IDH_UE_2013 + media_IDH_UE_2015)/3
media_IDH_Europeus= (media_IDH_Europeus_2010 + media_IDH_Europeus_2013 + media_IDH_Europeus_2015)/3
media_IDH_NE = (media_IDH_NE_2010 + media_IDH_NE_2013 + media_IDH_NE_2015)/3

medias_IDH = {'União Européia': media_IDH_UE , 'Europeus não participantes': media_IDH_Europeus , 'Não Europeus':media_IDH_NE}
names=list(medias_IDH.keys())
values=list(medias_IDH.values())

plt.bar(names,values, color = ('orange','b','g'))
plt.title("IDH")
plt.ylabel('Índice de Desenvolvimento Humano')
plt.show()
```

O gráfico acima evidencia a média do IDH de cada grupo ao longo do tempo. Evidentemente, os países da UE têm uma média maior que os outros, seguido pelos países europeus não participantes. Apesar da diferença aparentar ser sutil, deve se manter em mente que uma variação pequena no IDH já afeta muito as interpretações que ele oferece. Enquanto os países europeus não participantes apresentam 7.5% a menos dos da UE, os não europeus têm 25%. Portanto, pode se evidenciar que mesmo os países europeus tendo uma tendência elevada nesse quesito, os países na UE se destacam novamente, como nos outros índices.

```python
anos= [2010,2013,2015]
medias_IDH_UE=[media_IDH_UE_2010, media_IDH_UE_2013, media_IDH_UE_2015]
medias_IDH_Europeus=[media_IDH_Europeus_2010, media_IDH_Europeus_2013, media_IDH_Europeus_2015]
medias_IDH_NE=[media_IDH_NE_2010, media_IDH_NE_2013, media_IDH_NE_2015]

plt.scatter(anos,medias_IDH_UE, label = 'União Européia', color= 'orange')
z = np.polyfit(anos, medias_IDH_UE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"--", color = 'orange')


plt.scatter(anos,medias_IDH_Europeus, label = 'Europeu não participante', color = 'b')
z = np.polyfit(anos, medias_IDH_Europeus, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"b--")


plt.scatter(anos,medias_IDH_NE, label = 'Não Europeu', color = 'g')
z = np.polyfit(anos, medias_IDH_NE, 1)
p = np.poly1d(z)
plt.plot(anos,p(anos),"g--")


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Índice de Desenvolvimento Humano')
plt.xlabel('Anos')
plt.title('IDH')
plt.grid(True)
plt.show()
```

Por fim, sustentando a análise acima, este scatterplot mostra claramente as diferenças com relação à média do IDH dos países europeus, ao que eles se comportam em um patamar elevado quando comparados aos países não europeus. Ainda assim, os países da UE apresentam uma média distinta, apesar do patamar elevado já definido pelos países europeus. Esse fenômeno pode ser consequência da melhor condição econômica apresentada pelos outros indicadores acima, que muitas vezes facilita com que um país tenha recursos para melhorar a expectativa de vida, alfabetização, e poder de compra de sua população, além de vários outros focos sociais, fazendo com que o seu IDH médio seja mais alto. Além disso, todos os grupos apresentam uma taxa de variação positiva ao longo dos anos analisados, mostrando uma tendência mundial positiva com relação ao IDH, possivelmente devido aos avanços tecnológicos, econômicos e socias, além da conscientização social, que impactam muito a tendência do IDH ao longo do tempo.

# Conclusão:

A análise do PIB per capita demonstrou que os países da UE não somente têm valores maiores, mas que também crescem à uma taxa maior. Isso significa que os países da UE têm uma produção relativamente elevada, e assim proporcionam um PIB per capita maior, algo que reflete uma economia saudável e vantajosa. 

Já a análise do investimento estrangeiro demonstrou como que os países da UE costumavam a ter valores mais altos, mas que, juntos com os outros grupos, vem caindo bastante. Isso pode, no entanto, ser positivo ou negativo, ao que uma das perspectivas aponta que o fato de que os países estão passando a funcionar de forma mais autônoma, sem necessidade de capital estrangeiro, é uma coisa muito boa, enquanto em contrapartida isso pode ser consequência da falta de confiança do investidor estrangeiro. 

O nível de exportações mostrou algo muito positivo, uma vez que a média da UE não somente é a maior (por uma larga margem), mas ela é a que tem a maior taxa de variação. Isso pode refletir uma fonte de capital bastante forte na economia desses países, algo que é muito positivo para as suas situações econômicas. 

E por fim, os índices mencionados acima refletem, de certa forma, no desempenho extraordinário da média do IDH dos países na UE, que é bastante maior que a dos países não europeus, e ainda um tanto maior que a dos europeus não participantes.

Dessa forma, a União Européia aparenta ter de fato um impacto na economia de seus países ao que, por exemplo, os níveis extraordinários de exportações são em parte devido à coesão econômica mencionada nos objetivos da UE, que facilita o comércio entre os países participantes. Os indicadores utilizados nessa análise mostram, em maioria, que a União Européia tem tendências econômicas superiores às dos outros grupos, algo que aponta para que a resposta da pergunta proposta no início do projeto seja de que a UE cumpre os seus objetivos econômicos, e de que existe, de fato, uma vantagem econômica por estar na União Européia, com relação aos outros países europeus e todos os outros países no mundo.
